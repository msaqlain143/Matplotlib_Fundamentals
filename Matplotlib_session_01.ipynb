## Creating a simple plot

```python
# Importing NumPy for array manipulation
import numpy as np

# Importing Matplotlib in a Jupyter Notebook
import matplotlib.pyplot as plt

# This line is necessary to display plots directly in the notebook
%matplotlib inline

# Creating NumPy arrays for the example
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 6, 8, 10])

# Creating a simple line plot
plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line Plot')
plt.show()
```

## Line plots

```python
# Generating NumPy arrays for the example
x = np.linspace(0, 2*np.pi, 100)  # 100 points evenly spaced between 0 and 2*pi
y = np.sin(x)  # Computing the sine of each element in x

# Creating a simple line plot
plt.plot(x, y, label='sin(x)')  # Plotting y against x with a label
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line Plot of sin(x)')
plt.legend()  # Displaying the legend
plt.grid(True)  # Adding a grid for better readability
plt.show()
```

## Customizing line plots

```python
# Generating NumPy arrays for the example
x = np.linspace(-2*np.pi, 2*np.pi, 200)  # 200 points evenly spaced between -2*pi and 2*pi
y1 = np.sin(x)
y2 = np.cos(x)

# Creating a simple line plot with multiple lines
plt.plot(x, y1, label='sin(x)', color='blue', linestyle='--', linewidth=2)
plt.plot(x, y2, label='cos(x)', color='red', linestyle='-', linewidth=2)

# Adding labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot of sin(x) and cos(x)')

# Adding legend
plt.legend()

# Adding grid
plt.grid(True)

# Displaying the plot
plt.show()
```

```python
# Generating NumPy arrays for the example
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Creating a plot with customization options
plt.plot(x, y1, label='sin(x)', color='blue', linestyle='--', linewidth=2, marker='o', markersize=8)
plt.plot(x, y2, label='cos(x)', color='red', linestyle='-', linewidth=2, marker='s', markersize=8)

# Adding labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Customized Line Plot of sin(x) and cos(x)')

# Adding legend
plt.legend()

# Displaying the plot
plt.show()
```

## Scatter plots

```python
# Generating NumPy arrays for the example
x = np.random.rand(50)  # 50 random x-values between 0 and 1
y = np.random.rand(50)  # 50 random y-values between 0 and 1
#sizes = np.random.randint(10, 100, 50)  # 50 random sizes between 10 and 100

# Creating a scatter plot with customization options
#plt.scatter(x, y, s=sizes, c='green', marker='o', alpha=0.7, edgecolors='black', linewidths=1.5, label='Random Data')
plt.scatter(x, y, c='green', marker='o', alpha=0.7, edgecolors='black', linewidths=1.5, label='Random Data')
# Adding labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot with Customization')

# Adding legend
plt.legend()

# Displaying the plot
plt.show()
```

## Mixing scatter and line plots

```python
# Generating NumPy arrays for the example
x = np.linspace(0, 2*np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
scatter_x = np.linspace(0, 2*np.pi, 20)
scatter_y = np.sin(scatter_x) + 0.2 * np.random.randn(20)  # Adding some random noise for variety

# Creating a plot with a line plot and a scatter plot
plt.plot(x, y1, label='sin(x)', color='blue', linestyle='-', linewidth=2)
plt.plot(x, y2, label='cos(x)', color='red', linestyle='--', linewidth=2)
plt.scatter(scatter_x, scatter_y, label='Scatter Data', color='green', marker='o', s=50, alpha=0.8, edgecolors='black')

# Adding labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Combined Line and Scatter Plots')

# Adding legend
plt.legend()

# Displaying the plot
plt.show()
```

## Bar plots

```python
# Generating sample data for the example
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [25, 40, 30, 35]

# Creating a vertical bar plot
plt.bar(categories, values, color='skyblue')

# Adding labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Vertical Bar Plot')

# Displaying the plot
plt.show()
```

```python
# Generating sample data for the example
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [25, 40, 30, 35]

# Creating a horizontal bar plot
plt.barh(categories, values, color='lightcoral')

# Adding labels and title
plt.xlabel('Values')
plt.ylabel('Categories')
plt.title('Horizontal Bar Plot')

# Displaying the plot
plt.show()
```

## Histograms

```python
# Generating a random dataset for the example
data = np.random.randn(1000)  # 1000 random data points from a standard normal distribution

# Creating a histogram
plt.hist(data, bins=10, color='skyblue', edgecolor='black', alpha=0.7)

# Adding labels and title
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram Example')

# Displaying the plot
plt.show()
```

## Pie plots

```python
# Sample data for the example
sizes = [30, 25, 20, 15, 10]  # Sizes of different parts of the whole
labels = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']  # Labels for each part
colors = ['skyblue', 'lightcoral', 'lightgreen', 'gold', 'lightsalmon']  # Colors for each part
explode = (0.05, 0, 0, 0, 0)  # Exploding the first slice

# Creating a pie chart
plt.pie(sizes, labels=labels, colors=colors, explode=explode, autopct='%1.1f%%', startangle=90)

# Adding a title
plt.title('Pie Chart Example')

# Displaying the plot
plt.show()
```
